# Partie 1 Setup + index template 

## connexion elasticsearch

```python
from elasticsearch import Elasticsearch

es = Elasticsearch("http://localhost:9200")

print(es.info())
```

## Nettoyage 

```python
if es.indices.exists(index="library-books"):
    es.indices.delete(index="library-books")
```

## Index Template
on cée un template pour tous les index library-*

```python
template  = {
    "index_patterns": ["library-*"],
    "template": {
        "settings": {
            "number_of_shards": 1
        },
        "mappings": {
            "properties": {
                "title": {"type": "text", "fields": {"keyword": {"type": "keyword"}}},
                "author": {"type": "keyword"},
                "description": {"type": "text"},
                "category": {"type": "keyword"},
                "price": {"type": "float"},
                "published_date": {"type": "date"},
                "available": {"type": "boolean"},
                "rating": {"type": "float"}
            }
        }
    }
}

es.indices.put_index_template(name="library_template", body=template)
print("Template créé")
```

## Création index + alias

```python
# Supprime l'index s'il existe
if es.indices.exists(index="library-books-v1"):
    es.indices.delete(index="library-books-v1")

es.indices.create(
    index="library-books-v1",
    body={
        "aliases": {
            "books": {}
        }
    }
)
```

# Partie 2 (crud + search avancé)

## create/ajouter in livre (CRUD)

```python
doc = {
    "title": "Deep Learning guide",
    "author": "Alice Cooper",
    "description": "comprehensive deep learning book",
    "category": "ai",
    "price": 60,
    "published_date": "2023-08-01",
    "available": True,
    "rating": 4.0
}

res = es.index(index="books", document=doc)
print("Livre ajouté avec ID : ", res["_id"])
```

## read / récupérer un livre

```python
book_id = res["_id"]
book = es.get(index="books", id=book_id)
print(book["_source"])
```

## Update / modifier un livre

```python
es.update(
    index="books",
    id=book_id,
    doc={"price": 55, "rating": 5.0}
)

# verification
updated_book = es.get(index="books", id=book_id)
print(updated_book["_source"])
```

## Delete / Supprimer un livre

```python
es.delete(index="books", id=book_id)

#verifier
try:
    es.get(index="books", id=book_id)
except:
    print("Livre supprimé")
```

## Recherche full-text (title + description)

```python
res = es.search(
    index="books", 
    query={
        "multi_match": {
            "query": "machine learning",
            "fields": ["title", "description"]
        }
    }
)

print("Resultats recherche full-text : ")
for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

## Recherche filtrée  / avancée

### Exemple: AI books under $55

```python
res = es.search(
    index="books", 
    query={
        "bool": {
            "must": [
                {"term": {"category": "ai"}}
            ],

            "filter": [
                {"range": {"price": {"lte": 55}}},
                {"term": {"available": True}}
            ]
        }
    }
)

print("Filtres avancés : ")
for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

## Tri et pagination

```python
res = es.search(
    index="books", 
    query={"match_all": {}},
    sort=[{"price": "asc"}],
    from_=0,
    size=5
)

print("Top 5 livres triés par prix")
for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

## Recherche par alias (versioning avancé)

```python
# Alias "books" pointant sur library-books-v1 
res = es.search(index="books", query={"match_all": {}})
print("Tous les livres via alias : ")

for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

## Bonus : multi field (text + keyword)

```python
#Rechercher "python" dans title et trier exact par title.keyword
res = es.search(
    index="books",
    query={"match": {"title": "Python"}},
    sort=[{"title.keyword": "asc"}]
)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

# Nested Relations + orders + reviews

# Index Orders (Nested)

Chaque commande contient plusieurs livres -> nested

```python
# Supprime l'index s'il existe
if es.indices.exists(index="library-orders-v1"):
    es.indices.delete(index="library-orders-v1")

es.indices.create(
    index="library-orders-v1",
    body={
        "aliases": {
            "orders": {
                "is_write_index": True
            }
        },
        "mappings": {
            "properties": {
                "customer": {"type": "keyword"},
                "order_date": {"type": "date"},
                "total": {"type": "float"},
                "status": {"type": "keyword"},
                "items": {
                    "type": "nested",
                    "properties": {
                        "book_title": {
                            "type": "text",
                            "fields": {
                                "keyword": { "type": "keyword" }
                            }
                        },
                        "price": {"type": "float"},
                        "quantity": {"type": "integer"}
                    }
                }
            }
        }
    }
)

print("Index oreders créé")
```

## Recherche nested

```python
res = es.search(
    index="orders",
    query={
        "nested": {
            "path": "items",
            "query": {
                "match": {
                    "items.book_title": "Python Programming"
                }
            }
        }
    }
)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
```

## Aggregation top livres vedus

```python
res = es.search(
    index="orders", 
    size=0,
    aggs={
        "items_nested": {
            "nested": {"path": "items"},
            "aggs": {
                "top_books": {
                    "terms": {
                        "field": "items.book_title.keyword"
                    },
                    "aggs": {
                        "total_qty": {
                            "sum": {"field": "items.quantity"}
                        }
                    }
                }
            }
        }
    }
)

for bucket in res["aggregations"]["items_nested"]["top_books"]["buckets"]:
    print(bucket["key"], " -> quantité : ", bucket["total_qty"]["value"])
```

## Analyse clients (business)

```python
res = es.search(
    index="orders", 
    size=0,
    aggs={
        "clients": {
            "terms": {"field": "customer"},
            "aggs": {
                "total_spent": {
                    "sum": {"field": "total"}
                }
            }
        }
    }
)

for c in res["aggregations"]["clients"]["buckets"]:
    print(c["key"], "-> total: ", c["total_spent"]["value"])
```

## filtrage avancé (bool + nested)

```python
res = es.search(
    index = "orders",
    query = {
        "bool": {
            "must": [
                {"term": {"status": "paid"}}
            ],
            "filter": [
                {
                    "range": {
                        "total": {"gte": 50}
                    }
                }
            ]
        }
    }
)

for hit in res["hits"]["hits"]:
    print(hit["_source"])
```
